# Assignment 8 (Part A): Flight Information Visualization
## Aim: Visualize the data using Python libraries matplotlib, seaborn by plotting the graphs for assignment no. 3 (Flight Information System).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## 1. Load Data
Since the full 2008.csv is large, we demonstrate with a sample dataset structure.

```python
# Creating a sample representation of the Flight dataset
data = {
    'FlightNo': np.arange(1, 71),
    'Passengers': np.random.randint(50, 300, 70),
    'Delay': np.random.randint(-10, 120, 70),
    'Carrier': ['WN', 'AA', 'UA', 'DL', 'OO'] * 14,
    'Distance': np.random.randint(100, 2000, 70)
}
df = pd.DataFrame(data)
df.head()
```

## 2. Average Departure Delay by Day of Week

```python
df.columns
```

```python
plt.figure(figsize=(10,6))

sns.barplot(x='Carrier', y='Delay', data=df, estimator=np.mean)

plt.title('Average Delay by Carrier')

plt.ylabel('Delay (minutes)')

plt.show()
```

## 3. Departure Delay vs Arrival Delay

```python
plt.figure(figsize=(10,6))
sns.regplot(x='Distance', y='Delay', data=df)
plt.title('Departure Delay vs Arrival Delay')
plt.show()
```

## 4. Delay Distribution by Carrier

```python
plt.figure(figsize=(12,6))
sns.regplot(x='Distance', y='Delay', data=df)
plt.title('Arrival Delay Distribution per Carrier')
plt.show()
```
