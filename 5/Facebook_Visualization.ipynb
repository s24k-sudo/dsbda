# Assignment 5: Facebook Metrics Visualization
## Aim: Visualize the data using Python libraries matplotlib, seaborn by plotting the graphs for assignment no. 2.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../2/dataset_Facebook.csv', sep=';')
```

## 1. Bar Plot of Post Types

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Type', data=df, palette='viridis')
plt.title('Distribution of Post Types')
plt.show()
```

## 2. Scatter Plot: Likes vs Comments

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='like', y='comment', hue='Type', data=df)
plt.title('Likes vs Comments by Post Type')
plt.show()
```

## 3. Box Plot: Total Interactions by Category

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Category', y='Total Interactions', data=df)
plt.title('Total Interactions per Category')
plt.show()
```
